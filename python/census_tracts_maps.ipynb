```python
import pandas as pd
import geopandas as gpd
```

```python
# read in census tract data into geopandas dataframe

tracts_2000 = gpd.read_file('C:\\Users\\abonc\\OneDrive\\Documents\\GitHub\\urban_data_project\\inbound\\2000 Census Tracts.zip')
tracts_2010 = gpd.read_file('C:\\Users\\abonc\\OneDrive\\Documents\\GitHub\\urban_data_project\\inbound\\2010 Census Tracts.zip')
tracts_2020 = gpd.read_file('C:\\Users\\abonc\\OneDrive\\Documents\\GitHub\\urban_data_project\\inbound\\2020 Census Tracts.zip')
```

```python
tracts_2020
```

```python
# create dataframes for census tracts for 2000, 2010, and 2020
ct_2000 = tracts_2000[['boro_ct200']].sort_values(by='boro_ct200')
ct_2010 = tracts_2010[['boro_ct201']].sort_values(by='boro_ct201')
ct_2020 = tracts_2020[['boroct2020']].sort_values(by='boroct2020')
```

```python
print("Census tracts in 2000:", len(ct_2000))
print("Census tracts in 2010:", len(ct_2010))
print("Census tracts in 2020:", len(ct_2020))
```

```python
# merge dataframes together to see census tract changes over the years
df_ct_2010_to_2020 = ct_2020.merge(ct_2010,left_on='boroct2020', right_on='boro_ct201', how='outer')
df_ct_2000_to_2010 = ct_2000.merge(ct_2010,left_on='boro_ct200', right_on='boro_ct201', how='outer')

df_ct_2010_to_2020
```
